````python
# folium 라이브러리 추가
import folium
````

#### 지도 생성
- folium.Map() 클래스 생성자 실행(대문자 M/소문자m은 함수임)
    - 파라미터 location=[위도,경도]
    - zoom_start=1~20(1 전체지도표시, 20 최대확대)

```python
folium.Map(location=[35.229499, 129.090613], zoom_start=15)
```

````python
m=folium.Map(location=[35.229499, 129.090613], zoom_start=15)
m
````

#### 맵형태 변경
기본값 : OpenStreetMap
옵션 : stamenterrain, stamentoner, stamenwatercolor, cartodbpositron, cartodbdark_matter

````python
m2=folium.Map(location=[35.229499, 129.090613], zoom_start=15, tiles='stamentoner')
m2
````

#### 마커표시
```python
folium.Marker(location=[위도,경도], popup='팝업명', [icon=folium.Icon(color='색상', icon='부트스트랩 아이콘명'])).add_to(folium객체)     ##[icon~아이콘명']은 optional
```

````python
m=folium.Map(location=[35.256989, 127.810214], zoom_start=6)
folium.Marker(location=[35.229499, 129.090613], popup='한가람IT').add_to(m)    #팝업은 클릭해야 뜸
folium.Marker(location=[35.233846, 129.079539], popup='부산대학교', \
                icon=folium.Icon(color='red', icon='ok-circle')).add_to(m)
m
````

````python
## 지도에 구역 경계표시
import json
m=folium.Map(location=[35.256989, 127.810214], zoom_start=6)
gfile=open('./skorea-municipalities-2018-geo.json','r',encoding='utf-8').read()
jsondata=json.loads(gfile)
folium.GeoJson(jsondata, name='municipalities').add_to(m)
m
````
